### 5

Note that the `DebtRatio` column contains implausible data. Only the values ​​corresponding to known monthly income are ratios. The rest are absolute values ​​of monthly interest payments.

Correct the data by making all values ​​in the `DebtRatio` column absolute (multiply them by `MonthlyIncome`). To ensure that your program runs quickly on complete data, try to avoid using a loop.

#### *Functions that may be useful when solving:*

Accessing DataFrame elements:
* element: `data.loc[i, 'columnname']`
* column: `data['columnname']`
* submatrix: `data.loc[a:b, listColumnNames]`

Conditional indexing:
* `data.loc[data['column'] > 20, listColumnNames]`

It is better to write it like this:

* `i = data['column'] > 20` # vector of `True` and `False`
* `data.loc[i, 'columnname']`

Submatrices inherit their row numbers from the original.

* `pandas.isnull(scalar or array)` - check if value is undefined (`NaN`)
* `pandas.notnull(scalar or array)` - check if value is defined (not `NaN`)

```python
# место для кода
df['DebtRatio'] = df['DebtRatio'] * df['MonthlyIncome']
df
```

### 7

Calculate the average monthly income and assign all customers with unknown income to this number.

### 9c*
Visualize pairwise dependencies between non-binary features `'age', 'MonthlyIncome', 'NumberOfDependents'`. Limit monthly income to 25000.

What patterns can you observe in the resulting graphs?

*Functions that may be useful in solving: `pd.plotting.scatter_matrix()`*

```python
import seaborn as sns

sns.kdeplot(df[df['SeriousDlqin2yrs'] == 1]['MonthlyIncome'], color='red', label='клиентов с задолженностями')
sns.kdeplot(df[df['SeriousDlqin2yrs'] == 0]['MonthlyIncome'], color='blue', label='клиентов без задолженностей')

plt.xlim(0, 25000)
plt.xlabel('месячный доход')
plt.ylabel('плотность')
plt.legend()

plt.show()
```

```python
filtered_df = df[df['MonthlyIncome'] <= 25000]

features = ['age', 'MonthlyIncome', 'NumberOfDependents']

pd.plotting.scatter_matrix(filtered_df[features], figsize=(10, 10), alpha=0.7, diagonal='kde')

plt.show()
```
